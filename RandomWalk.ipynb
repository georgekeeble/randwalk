### Abstract/Motivation

We see examples of random walks in day to day life, for example, whilst adding a drop of food colouring to a glass of water will depend partially on any potential current in the water it will also depend on a **random walk**. Interestingly the way that a random walk behaves in 2D and 3D space is goverened in the same way as the elementary 1D **Random Walk**. other examples include a gas being released in a room and many biological movement models are based on extensions of a **simple random walk**.
Below we shall explore **Random Walks** in some detail looking at methods to ensure we have a truly random path and then we shall go on to look at the **Mean Square Displacement (MSD)** for a given trajectory.
```python
import numpy as np 
from math import *
import matplotlib.pyplot as plt 

#Function to simulate a random walk 

def randwalk(n):

    X = np.zeros(n)
    Y = np.zeros(n)
    Thetas = 2*pi*np.random.rand(n)
    RW = np.zeros((n,2))
    for i in range(n-1):
        X[i+1]= X[i] + cos(Thetas[i])
        Y[i+1] = Y[i] + sin(Thetas[i])
        RW[i+1]= (X[i+1],Y[i+1])
    return RW
plt.plot(randwalk(10000)[:,0],randwalk(10000)[:,1])
```

```python
#Now for average random walk 
#Want m lots of simulations of a walk of n increments 


def AVE_randwalk(n,m):
    MAT_X = np.zeros((n,m))
    MAT_Y = np.zeros((n,m))
    AV_X= np.zeros(n)
    AV_Y=np.zeros(n)
    for i in range(m):
        MAT_X[:,i] = randwalk(n)[:,0]
        MAT_Y[:,i] = randwalk(n)[:,1]
    for j in range(n):
        AV_X[j]=np.sum(MAT_X[j,:])
        AV_Y[j]=np.sum(MAT_Y[j,:])
    
    return plt.plot(AV_X,AV_Y)
        
        
        
    
AVE_randwalk(10000,30)
```
